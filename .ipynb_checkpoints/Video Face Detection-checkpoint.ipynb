## Face detection using OpenCV

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
# converting BGR to RBG format
def convertToRGB(img):
    return cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
# Loading the classifer for frontal face
haar_cascade_face = cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')
```

#### Face detection
    We use detectMultiscale module of the classifer. This function returns a rectangle with coordinates (x,y,w,h) around the detected faces.
    This function has two important parameters which have to be tuned according to the data.
    - scalefactor: In a group of photo, there may be some faces which are near the camera than others. naturally, such faces would appear more prominent than the one behind. This factor compensates for that.
    
    - minNeighbours: This parameter specifies the number of neighbours a rectangle should have to be called a face.

```python
def detect_faces(cascade, image, scaleFactor=1.1):
    # create a copy of the image to prevent any changes to the original image.
    img_copy = image.copy()
    
    # convert the image to gray scale as opencv detector expects gray scale image.
    gray_scale = cv2.cvtColor(img_copy,cv2.COLOR_BGR2GRAY)
    
    # Applying the haar classifier to detect faces
    faces_react = cascade.detectMultiScale(gray_scale, scaleFactor=scaleFactor, minNeighbors=5)
        
    for i,(x,y,w,h) in enumerate(faces_react):
        cv2.rectangle(img_copy,(x,y),(x+w,y+h-10),(0,0,255),2)
        font = cv2.FONT_HERSHEY_SIMPLEX
        line_type = cv2.LINE_AA
        font_color = (0,0,255)
        if i == (len(faces_react)-1):
            font_color = (0,255,0) 
        cv2.putText(img_copy,'Person '+str(i), (x,y+h),font,0.35,font_color,1,line_type)
    
    return img_copy
```

```python
# testing for a static image
sample_img = cv2.imread('./images/group1.jpg')
faces = detect_faces(haar_cascade_face,sample_img,1.1)
plt.imshow(convertToRGB(faces))
```

```python
# get web cam feed
webcam = cv2.VideoCapture(0)
while True:
    (_,im) = webcam.read()    
    faces = detect_faces(haar_cascade_face,im,1.1)
#     color_img = convertToRGB(faces)
    cv2.imshow('OpenCV Face Recognition (Press esc to exit)',faces)
    key = cv2.waitKey(10)
    if key == 27:
        break
webcam.release()
cv2.destroyAllWindows()
```
